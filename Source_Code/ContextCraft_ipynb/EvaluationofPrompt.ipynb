```python
import pandas as pd
# Load the dataset into a DataFrame
# Replace 'your_dataset.csv' with the path to your actual dataset file
file_path = '/content/drive/MyDrive/Method/EnglishMethod/PromptProcessDataset/Results/result_final.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to ensure it's loaded correctly
print(df)
```

```python
# Get the first 151 lines and assign to a new DataFrame
dfr = df

# Display the first few rows of the new DataFrame to ensure it's loaded correctly
dfr.head()
```

```python
# Calculate the number of words in the "Functional Description" column
dfr['Word_Count'] = dfr['Functional Description'].apply(lambda x: len(str(x).split()))

# Find the maximum word count
max_word_count = dfr['Word_Count'].max()

# Define bins: very short (0-5), intermediate groups, very long
very_short_threshold = 5
very_long_threshold = max_word_count - (max_word_count // 10)

# Ensure there are at least 10 bins in total
intermediate_step = (very_long_threshold - very_short_threshold) // 8

bins = [0, very_short_threshold] + [very_short_threshold + i * intermediate_step for i in range(1, 9)] + [float('inf')]
labels = ['0-5'] + [f'{very_short_threshold + (i - 1) * intermediate_step + 1}-{very_short_threshold + i * intermediate_step}' for i in range(1, 9)] + [f'{very_long_threshold + 1}+']

# Create a new column for word count category
dfr['Word_Count_Category'] = pd.cut(dfr['Word_Count'], bins=bins, labels=labels, right=False)

# Split the dataset into different groups based on word count category
word_count_groups = {label: dfr[dfr['Word_Count_Category'] == label] for label in labels}

# Display the DataFrame with the new columns for verification
print(dfr)
```

```python
import pandas as pd
import Levenshtein
import re
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Function to calculate raw Levenshtein distance
def levenshtein_distance(actual, predicted):
    # Ensure both inputs are strings and strip whitespace
    actual = str(actual).strip()
    predicted = str(predicted).strip()
    actual = to_camel_case(actual)
    predicted = to_camel_case(predicted)
    return Levenshtein.distance(actual, predicted)

# Function to calculate exact match
def exact_match(actual, predicted):
    # Ensure both inputs are strings and strip whitespace
    actual = str(actual).strip()
    predicted = str(predicted).strip()
    actual = to_camel_case(actual)
    predicted = to_camel_case(predicted)
    return 1 if actual == predicted else 0

# Function to convert camel case to sentence
def camel_case_to_sentence(name):
    s = re.sub('([a-z0-9])([A-Z])', r'\1 \2', name)
    return s
def to_camel_case(method_name):
    # Change the first character to lowercase if it is uppercase
    if method_name and method_name[0].isupper():
        return method_name[0].lower() + method_name[1:]
    return method_name


# Load the pre-trained model for semantic similarity
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
# Function to calculate semantic similarity
def semantic_similarity(actual, predicted):
    actual = str(actual).strip()
    predicted = str(predicted).strip()
    actual_sentence = camel_case_to_sentence(actual)
    predicted_sentence = camel_case_to_sentence(predicted)
    actual_embedding = model.encode([actual_sentence])
    predicted_embedding = model.encode([predicted_sentence])
    similarity = cosine_similarity(actual_embedding, predicted_embedding)
    return similarity[0][0]


# List of columns containing predicted method names
predicted_columns = ['ChatGPT3.5_T', 'ChatGPT4_T', 'ChatGPT4o_T', 'ChatGPT-4o-zero', 'ChatGPT4o-Few',
                     'ChatGTP3.5_Zero', 'ChatGPT3.5_Few', 'ChatGPT4_Zero', 'ChatGPT4_Few',
                     'Gemini_Zero', 'Gemini_Few', 'Gemini_T', 'Llama3-zero', 'Llama3-few', 'Llama3_T']
```

```python
import pandas as pd

# Apply the calculations for each DataFrame in word_count_groups
all_results = []

for label, df_group in word_count_groups.items():
    for col in predicted_columns:
        # Calculate Levenshtein distance for each row and store it in a new column
        df_group[col + '_Edit_Distance'] = df_group.apply(lambda row: levenshtein_distance(row['Method Name'], row[col]), axis=1)
        # Calculate exact match for each row and store it in a new column
        df_group[col + '_Exact_Match'] = df_group.apply(lambda row: exact_match(row['Method Name'], row[col]), axis=1)
        # Calculate semantic similarity for each row and store it in a new column
        df_group[col + '_Semantic_Similarity'] = df_group.apply(lambda row: semantic_similarity(row['Method Name'], row[col]), axis=1)

    # Add a column for the group label
    df_group['Group'] = label

    # Append the modified group DataFrame to the list
    all_results.append(df_group)

# Concatenate all individual results into a single DataFrame
final_results_df = pd.concat(all_results)

# Display the final results DataFrame
print(final_results_df)

# Optional: Save the final results DataFrame to a CSV file
final_results_df.to_csv('final_results.csv', index=False)
```

```python
# Function to split camel case into tokens
def split_camel_case(name):
    return re.findall(r'[A-Z](?:[a-z]+|[A-Z]*(?=[A-Z]|$))', name)

# Function to count tokens and assign label
def count_tokens_and_label(name):
    tokens = split_camel_case(name)
    num_tokens = len(tokens)
    if num_tokens == 1:
        return '1'
    elif num_tokens == 2:
        return '2'
    elif num_tokens == 3:
        return '3'
    elif num_tokens == 4:
        return '4'
    elif num_tokens == 5:
        return '5'
    else:
        return '6+'

# Apply the function to create the 'Token Label' column
df_coffi['Method Length'] = df_coffi['Method Name'].apply(count_tokens_and_label)

# Display the updated DataFrame
df_coffi.head()
```

```python
import pandas as pd
import numpy as np


df = df_coffi

# Convert 'Description Length' to numeric by taking the midpoint of each range
def midpoint(range_str):
    try:
        parts = range_str.split('-')
        return (int(parts[0]) + int(parts[1])) / 2
    except:
        return np.nan

df['Description Length (Numeric)'] = df['Description Length'].apply(midpoint)

# Ensure columns 'EM', 'SSI', 'ED' are numeric
df[['EM', 'SSI', 'ED']] = df[['EM', 'SSI', 'ED']].apply(pd.to_numeric, errors='coerce')

# Treat 'Description Length' and 'Method Length' as categorical directly
df['Method Length (Categorical)'] = df['Method Length']

# Calculate CC1: Aggregate means by 'Description Length' and compute correlations
mean_values_description = df.groupby('Description Length')['EM', 'SSI', 'ED'].mean().reset_index()
correlation_description_length_cat = mean_values_description.corr()

# Calculate CC2: Aggregate means by 'Method Length' and compute correlations
mean_values_method = df.groupby('Method Length (Categorical)')['EM', 'SSI', 'ED'].mean().reset_index()
correlation_method_length_cat = mean_values_method.corr()

# Display the results
correlation_description_length_cat, correlation_method_length_cat

import matplotlib.pyplot as plt

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Bar chart for CC1
correlation_description_length_cat.iloc[0][['EM', 'SSI', 'ED']].plot(kind='bar', ax=ax[0], color=['skyblue', 'orange', 'green'])
ax[0].set_title('Correlation Coefficients (CC1)\nAggregated Description Length vs EM, SSI, ED')
ax[0].set_ylabel('Correlation Coefficient')
ax[0].set_ylim(-1, 1)

# Bar chart for CC2
correlation_method_length_cat.iloc[0][['EM', 'SSI', 'ED']].plot(kind='bar', ax=ax[1], color=['skyblue', 'orange', 'green'])
ax[1].set_title('Correlation Coefficients (CC2)\nAggregated Method Length vs EM, SSI, ED')
ax[1].set_ylabel('Correlation Coefficient')
ax[1].set_ylim(-1, 1)

plt.tight_layout()
plt.show()

```

```python
# Calculate exact match, raw Levenshtein distance, and semantic similarity for each predicted column
edit_distance_scores = {}
exact_match_scores = {}
semantic_similarity_scores = {}

for col in predicted_columns:
    # Calculate Levenshtein distance for each row and store it in a new column
    dfr[col + '_Edit_Distance'] = dfr.apply(lambda row: levenshtein_distance(row['Method Name'], row[col]), axis=1)
    # Calculate exact match for each row and store it in a new column
    dfr[col + '_Exact_Match'] = dfr.apply(lambda row: exact_match(row['Method Name'], row[col]), axis=1)
    # Calculate semantic similarity for each row and store it in a new column
    dfr[col + '_Semantic_Similarity'] = dfr.apply(lambda row: semantic_similarity(row['Method Name'], row[col]), axis=1)

    # Compute the average Levenshtein distance for the current column
    avg_edit_distance = dfr[col + '_Edit_Distance'].mean()
    edit_distance_scores[col] = avg_edit_distance

    # Compute the average exact match for the current column
    avg_exact_match = dfr[col + '_Exact_Match'].mean()
    exact_match_scores[col] = avg_exact_match

    # Compute the average semantic similarity for the current column
    avg_semantic_similarity = dfr[col + '_Semantic_Similarity'].mean()
    semantic_similarity_scores[col] = avg_semantic_similarity

# Create a DataFrame to present the results
results_df = pd.DataFrame({
    'Model': predicted_columns,
    'Average Edit Distance': [edit_distance_scores[col] for col in predicted_columns],
    'Average Exact Match': [exact_match_scores[col] for col in predicted_columns],
    'Average Semantic Similarity': [semantic_similarity_scores[col] for col in predicted_columns]
})

# Display the results
print(results_df)
```
